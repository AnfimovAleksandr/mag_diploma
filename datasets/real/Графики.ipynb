# Данные по Старому Служаке

```python
# Импортируем необходимые библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем данные из CSV-файла
df = pd.read_csv('faithful.csv', index_col=0)

# Настраиваем стиль визуализации
sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(12, 8))

# Создаем scatter plot без регрессионной линии
scatter_plot = sns.scatterplot(
    data=df,
    x="eruptions",
    y="waiting",
    color="darkblue",
    alpha=0.7,
    s=60
)

# Добавляем заголовок и подписи осей
# plt.title('Зависимость между продолжительностью извержения и временем ожидания\nГейзер "Старый Служака" (Озеро Йеллоустон)', 
#           fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Продолжительность извержения (минуты)', fontsize=12)
plt.ylabel('Время ожидания до следующего извержения (минуты)', fontsize=12)

# Добавляем сетку для удобства чтения
plt.grid(True, linestyle='--', alpha=0.7)

# Добавляем аннотацию с количеством наблюдений
plt.annotate(f'Количество наблюдений: {len(df)}', 
             xy=(0.05, 0.95), xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

# Настраиваем отображение значений на осях
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Автоматическая подгонка layout для предотвращения обрезки элементов
plt.tight_layout()

# Сохраняем график в файл
plt.savefig('old_faithful_scatter.png', dpi=300, bbox_inches='tight')

# Отображаем график
plt.show()

# Дополнительно: строим гистограммы для анализа распределений
plt.figure(figsize=(14, 6))

# Гистограмма продолжительности извержений
plt.subplot(1, 2, 1)
sns.histplot(df['eruptions'], bins=15, kde=True, color='teal')
plt.title('Распределение продолжительности извержений', fontsize=14)
plt.xlabel('Продолжительность (минуты)', fontsize=12)
plt.ylabel('Частота', fontsize=12)

# Гистограмма времени ожидания
plt.subplot(1, 2, 2)
sns.histplot(df['waiting'], bins=15, kde=True, color='purple')
plt.title('Распределение времени ожидания', fontsize=14)
plt.xlabel('Время ожидания (минуты)', fontsize=12)
plt.ylabel('Частота', fontsize=12)

plt.tight_layout()
plt.savefig('old_faithful_distributions.png', dpi=300, bbox_inches='tight')
plt.show()
```
